###Question 2: What nationalities are being deported the most?
---
I will arrange my country_sums.csv set to see what nationality was the most represented in deported individuals through Secure Communities from 2015-2017.

```
import pandas as pd

country_sums = pd.read_csv('https://raw.githubusercontent.com/mgmills93/AEM_Spring_2019/master/SecureCommunities_FinalProject/Datasets/country_sums.csv')

def top_5(country_sums):
  country_sort = country_sums.sort_values(by=['Arrest Date'], ascending=False)
  country_sort = country_sort.reset_index(drop=True)
  
  return country_sort
country_sort = top_5(country_sums)
```

```
def clean(country_sort):
  country_sort.rename(columns={'Arrest Date':'Arrest_Number'}, inplace=True)
  final_sort = country_sort[['Citizenship Country', 'Arrest_Number']]
  
  return final_sort
```

###Pipeline


```
#sort country_sums by largest number of arrests
top_5(country_sums)
#clean sorted data
clean(country_sort)
```

###Analysis
---
The top 5 countries with the most nationals deported from the U.S. from 2015-2017 were:
Mexico - 113697  
Honduras -	9086  
Guatemala -	8610

Dominican Republic - 1800
