###Question 1: Is there a “Trump Effect” in SC arrests overall?
---
I will plot the dataset I created with arrests by month-year as a bar graph. 

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

trump_monthyear = pd.read_csv('https://raw.githubusercontent.com/mgmills93/AEM_Spring_2019/master/SecureCommunities_FinalProject/Datasets/trump_effect.csv')
my_xticks = ['2014-10', '2014-11', '2014-12', '2015-01', '2015-2', '2015-3', '2015-4','2015-5','2015-6','2015-7','2015-8','2015-9','2015-10','2015-11','2015-12', '2016-01', '2016-02','2016-03','2016-04','2016-05','2016-06','2016-07','2016-08','2016-09','2016-10','2016-11','2016-12', '2017-01', '2017-02', '2017-03']
trump_graph = trump_monthyear[['Month_Year', 'Departed Date']].plot(kind='bar', title="SC Arrests by Month", figsize=(15, 10), legend=True, fontsize=12)
trump_graph.set_xlabel('Month',fontsize=12)
plt.xticks(range(len(my_xticks)), my_xticks, size='small')
trump_graph.set_ylabel("# of Arrests",fontsize=12)


trump_fig = trump_graph.get_figure()
trump_fig.savefig("trump_fig.png")
```

###Analysis
---
November 2016 is labeled as '25' on the x-axis. The arrest value for this month is slightly higher than surrounding months, but it is not an outlier when comparing to the data as a whole. To answer my research question, there was not a "Trump effect", or spike in Secure Communities arrests in November 2016.
