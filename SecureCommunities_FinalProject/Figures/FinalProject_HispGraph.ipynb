```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
final_hisp_set = pd.read_csv('https://raw.githubusercontent.com/mgmills93/AEM_Spring_2019/master/SecureCommunities_FinalProject/Datasets/final_hisp_set.csv')
print(final_hisp_set.sample(5))
# plot
plt.plot( 'Percent_Hispanic', 'Arrest_Number', data=final_hisp_set, linestyle='none', marker='o')
plt.suptitle('Total County-Level Deportations & Hispanic Population Percentage')
plt.xlabel('% Hispanic')
plt.ylabel('County-Level Arrests')
axes = plt.gca()
axes.set_ylim([0,200])
plt.show()

plt.savefig('hisp_plot_zoom.png')
```

```
import seaborn as sns

hisp_plot = sns.regplot(final_hisp_set['Percent_Hispanic'], final_hisp_set['Arrest_Number'])
hisp_plot.set(xlabel='% Hispanic', ylabel='County-Level Deportations', title='Total County-Level Deportations & Hispanic Population Percentage')

fig = hisp_plot.get_figure()
fig.savefig('sns_hisp_plot.png')
```
